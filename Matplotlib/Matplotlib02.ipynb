```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
np_arr = np.linspace(0, 10, 20)
np_arr
```

```python
np_arr2 = np_arr ** 2
np_arr2
```

```python
(my_fig, my_ax) = plt.subplots()
my_ax.plot(np_arr, np_arr, color="blue" , alpha=0.5)
my_ax.plot(np_arr, np_arr2, color="#dee81a")

```

```python
(my_fig, my_ax)=plt.subplots()
my_ax.plot(np_arr, np_arr, color="#6865b5" , linewidth=4)
my_ax.plot(np_arr, np_arr+1, color="#dee81a", linewidth=5)

my_ax.plot(np_arr, np_arr+4, color="#6865b5", linewidth=3, linestyle="--")
my_ax.plot(np_arr, np_arr+5, color="#dee81a", linewidth=3, linestyle=":")
my_ax.plot(np_arr, np_arr+8, color="#6865b5", linewidth=3, linestyle="-.")

my_ax.plot(np_arr, np_arr+7, color="blue", linewidth=3, linestyle="--", marker="o", markersize=5, markerfacecolor="yellow")
```
