```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
age_list = [10, 20, 30, 30, 30, 40, 50, 60, 70, 75]
weight_list = [20, 60, 80, 85, 86, 87, 70, 90, 95, 90]
```

```python
numpy_age_list = np.array(age_list)
numpy_weight_list = np.array(weight_list)
```

```python
plt.plot(numpy_age_list, numpy_weight_list, "b", label="first")
plt.plot(numpy_age_list, numpy_weight_list+10, "r", label="second")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age-Weight Changing Graph")
plt.legend(loc=4)
#plt.show()
```

```python
numpy_array1 = np.linspace(1, 10, 20)
numpy_array1
```

```python
numpy_array2 = numpy_array1 ** 3
numpy_array2
```

```python
plt.subplot(1, 2, 1)
plt.plot(numpy_array1, numpy_array2, "g:")

plt.subplot(1, 2, 2)
plt.plot(numpy_array2, numpy_array1, "b*-")
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.2, 0.2, 0.8, 0.5])
figure_axes.plot(numpy_array1, numpy_array2, "b--")
figure_axes.set_xlabel("x-axis")
figure_axes.set_ylabel("y-axis")
figure_axes.set_title("Graphic title")
```

```python
figure2 = plt.figure()

axes_main = figure2.add_axes([0.1, 0.1, 0.8, 0.8])
axes_inner = figure2.add_axes([0.2, 0.6, 0.3, 0.2])

axes_main.plot(numpy_array1, numpy_array2, "b*-")
axes_main.set_xlabel("x-axis")
axes_main.set_ylabel("y-axis")
axes_main.set_title("Main Graph")

axes_inner.plot(numpy_array2, numpy_array1, "r*-")
axes_inner.set_xlabel("inner_x-axis")
axes_inner.set_ylabel("inner_y-axis")
axes_inner.set_title("Inner Graph")
```

# subplots

```python
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
```

```python
fig, axs = plt.subplots(2, 2)
fig.suptitle("Vertical stacked")
fig.dpi = 100
axs[0, 0].plot(x, y, "b")
axs[0, 0].set_xlabel("xxx")
axs[0, 1].plot(x, -y, "r")
plt.tight_layout()
```

# legend

```python
my_figure = plt.figure(dpi=100)
my_axes = my_figure.add_axes([0.1, 0.1, 0.9, 0.9])

my_axes.plot(numpy_array1, numpy_array1**2, "r*:", label="numpy_array**2")
my_axes.plot(numpy_array1, numpy_array1**3, "bo-.", label="numpy_array**3")
my_axes.legend(loc=2)
```
