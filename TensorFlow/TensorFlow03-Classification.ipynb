```python
y = dataset["Type"].values
x = dataset.drop("Type", axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.3, random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar = MinMaxScaler()
```

```python
scalar.fit(x_train)
```

```python
x_train = scalar.transform(x_train)
x_test = scalar.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
model=Sequential()

model.add(Dense(units=30, activation="relu"))

model.add(Dense(units=15, activation="relu"))
model.add(Dense(units=15, activation="relu"))

model.add(Dense(units=1, activation="sigmoid"))

model.compile(optimizer="adam", loss="binary_crossentropy")
```

```python
model.fit(x_train, y_train, epochs=700, validation_data=(x_test, y_test), verbose=1)
```

```python
model.history.history
```

```python
loss_df = pd.DataFrame(model.history.history)
```

```python
loss_df.plot()
```

# Early Stopping

```python
model = Sequential()

model.add(Dense(units=30, activation="relu"))

model.add(Dense(units=15, activation="relu"))
model.add(Dense(units=15, activation="relu"))

model.add(Dense(units=1, activation="sigmoid"))

model.compile(optimizer="adam", loss="binary_crossentropy")
```

```python
earlyStopping = EarlyStopping(monitor="val_loss", mode="min", patience=25, verbose=1)
```

```python
model.fit(x_train, y_train, epochs=700, validation_data=(x_test, y_test), verbose=1, callbacks=[earlyStopping])
```

```python
loss_df = pd.DataFrame(model.history.history)
```

```python
loss_df.plot()
```

# Dropout

```python
model = Sequential()

model.add(Dense(units=30, activation="relu"))
model.add(Dropout(0.5))

model.add(Dense(units=15, activation="relu"))
model.add(Dropout(0.5))
model.add(Dense(units=15, activation="relu"))
model.add(Dropout(0.5))

model.add(Dense(units=1, activation="sigmoid"))

model.compile(optimizer="adam", loss="binary_crossentropy")
```

```python
model.fit(x_train, y_train, epochs=700, validation_data=(x_test, y_test), verbose=1, callbacks=[earlyStopping])
```

```python
loss_df = pd.DataFrame(model.history.history)
```

```python
loss_df.plot()
```

```python
predictions = model.predict(x_test)

predictions = [1 if x >=0.5 else 0 for x in predictions]
print(predictions)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, predictions))
```

```python
confusion_matrix(y_test, predictions)
```
